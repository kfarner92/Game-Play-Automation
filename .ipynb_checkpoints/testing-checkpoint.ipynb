```python
#Dependencies
import pyautogui as pag
import numpy as np
import cv2
import math
from random import randint, uniform
import sys
import time
import os
```

```python
pag.PAUSE = 1.75
pag.FAILSAFE = True
pag.position()
pag.screenshot('images/full_client_screen.png',region=((5*2),(50*2),(960*2),(720*2)))
```

```python
triggers = (488,321,147,40,'images/existing_usr.png')
r = triggers[0]*2,triggers[1]*2,triggers[2]*2,triggers[3]*2
r_full =(5*2),(50*2),(960*2),(720*2)
pag.screenshot('images/screenshot.png', region=r_full)
screen = cv2.imread('images/screenshot.png',0)
template = cv2.imread(triggers[4],0) 
w, h = template.shape[::-1] 
res = cv2.matchTemplate(screen, template, cv2.TM_CCOEFF_NORMED)

threshold = .80
loc = np.where(res >= threshold)
if len(loc[0]) > 0:
    print(True)

else:
    print(False)
```
